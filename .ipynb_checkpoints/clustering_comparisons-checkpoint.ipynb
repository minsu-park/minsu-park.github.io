```python
from clusim.clustering import Clustering, print_clustering
import clusim.sim as sim

f_path =  '/Users/mp5500/Library/CloudStorage/Dropbox/cultural_clusters/'
```

```python
import pandas as pd

# Element-centeric Clustering Similarity
df = pd.read_csv(f_path+'cm_0.32.csv')
d_w_native = df.groupby('node')['cluster'].agg(list).to_dict()

df = pd.read_csv(f_path+'cm_0.36.csv')
d_w_scaled = df.groupby('node')['cluster'].agg(list).to_dict()
```

```python
def rename_keys(d, keys):
    return dict([(keys.get(k), v) for k, v in d.items()])

node_to_int_mapping = {}
for i, node in enumerate(df['node'].unique()):
    node_to_int_mapping[node] = i

d_w_native = rename_keys(d_w_native, node_to_int_mapping)
d_w_scaled = rename_keys(d_w_scaled, node_to_int_mapping)
```

```python
clustering_1 = Clustering(elm2clu_dict = d_w_native)
clustering_2 = Clustering(elm2clu_dict = d_w_scaled)

print_clustering(clustering_1)
print_clustering(clustering_2)
```

```python
print("Element-centric Similarity:", sim.element_sim(clustering_1, clustering_2, alpha = 0.9))

print("Element-centric Similarity for each element:")
elementScores, relabeled_elements = sim.element_sim_elscore(clustering_1, clustering_2, alpha = 0.9)
print(elementScores)
```

```python
import random

def random_choice_with_all_items(items, sample_size):
    """
    Returns a list of sample_size items from the list, making sure that all items appear at least once.

    Args:
    items: A list of items.
    sample_size: The number of items to sample.

    Returns:
    A list of sample_size items, with all items appearing at least once.
    """

    # Create a list of unique items.
    unique_items = list(set(items))

    # Sample the unique items until the sample size is reached.
    sampled_items = random.sample(unique_items, min(sample_size, len(unique_items)))

    # If the sample size is greater than the number of unique items, add additional samples from the unique items.
    if sample_size > len(unique_items):
        additional_samples = random.choices(unique_items, k=sample_size - len(unique_items))
        sampled_items.extend(additional_samples)

    return sampled_items
```

```python
import numpy as np

df = pd.read_csv(f_path+'cm_0.32.csv')
l_county = list(df['node'].unique())
df['node'] = random_choice_with_all_items(l_county, len(df['node']))
d_w_native_shuffled = df.groupby('node')['cluster'].agg(list).to_dict()

df = pd.read_csv(f_path+'cm_0.36.csv')
df['node'] = random_choice_with_all_items(l_county, len(df['node']))
d_w_scaled_shuffled = df.groupby('node')['cluster'].agg(list).to_dict()
```
